```python
import matplotlib
import matplotlib.pyplot as plt
from Bio import Phylo #pip3 install biopython

tree = Phylo.read('./sample/4_GP_cut_99_01G029800_fasta_guide_tree', 'newick')
matplotlib.rc('font', size = 6)
fig = plt.figure(figsize=(10, 15), dpi=100)
axes = fig.add_subplot(1, 1, 1)
Phylo.draw(tree, axes=axes)
#plt.savefig('./4_GP_cut_99_01G029200_fasta_tree.png', dpi=100)
```
