# Genbank file - matrix - make fasta file - phylogenetic tree(clustalw) (with definition fuction)

## From genbank file, make dataframe and fasta file with definition fuction

```python
#genbank file 불러올때 사용
from Bio import SeqIO
#fasta file만들때 사용
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import pandas as pd
#dataframe merge
from functools import reduce
```

```python
def make_CDS_dic(genbank_file_root) :
    gb_parse = [file for file in SeqIO.parse(genbank_file_root, "genbank")]
    
    CDS_dic = {}
    for gb in gb_parse :
        feats = [feat for feat in gb.features if feat.type == "CDS"]
        for feat in feats :
            key = feat.qualifiers['gene'][0]
            value = str(feat.location.extract(parent_sequence = gb.seq).translate(table=11, to_stop=True))
            CDS_dic[key] = value
        return(CDS_dic)
```

```python
def make_df(genbank_file_root) :
    gb_parse = [file for file in SeqIO.parse(genbank_file_root, "genbank")]
    
    name = genbank_file_root.split('/')[-1].split('.gb')[0]
    CDS_dic = {}
    for gb in gb_parse :
        feats = [feat for feat in gb.features if feat.type == "CDS"]
        for feat in feats :
            try :
                key = feat.qualifiers['gene'][0]
                value = str(feat.location.extract(parent_sequence = gb.seq).translate(table=11, to_stop=True))
                CDS_dic[key] = value
            except KeyError :
                pass
    df_CDS = pd.DataFrame(list(CDS_dic.items()), columns=['Gene name', '%s'%name])
    return(df_CDS)
```

```python
def draw_tree(root) :
    tree = Phylo.read(root, 'newick')
    matplotlib.rc('font', size = 6)
    fig = plt.figure(figsize=(5, 5), dpi=100)
    axes = fig.add_subplot(1, 1, 1)
    return(Phylo.draw(tree, axes=axes))
```
