### Get genbank format file without downloading

http://www.incodom.kr/E-utility

```
! pip3 install biopython
from Bio import Entrez
from Bio import SeqIO
```

```
Entrez.email = "lim@forest.com" #Email address
```

```
nodown_gb = Entrez.efetch(db="nucleotide", id="JN160605.1", rettype="gb", retmode="text")
```

```
print(nodown_gb.read())
```

```
# 위에서 read로 한 번 읽었으니 다시 읽어야 열 수 있음

nodown_gb = Entrez.efetch(db="nucleotide", id="JN160605.1", rettype="gb", retmode="text")
```

```
nodown_gb_parse = [file for file in SeqIO.parse(nodown_gb, 'genbank')]
```

```
print(len(nodown_gb_parse))
```

```
print(nodown_gb_parse)
```

```
for nodown_gb in nodown_gb_parse :
  print(nodown_gb.annotations)
```
